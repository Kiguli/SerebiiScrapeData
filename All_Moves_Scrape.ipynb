# Preliminaries

```python
import bs4 as bs
import urllib.request
import pickle
import pandas as pd
import numpy as np
```

# Physical Moves

```python
#website to scrape
sauce = urllib.request.urlopen('https://www.serebii.net/attackdex-dp/physical.shtml').read()
soup = bs.BeautifulSoup(sauce,'lxml')
```

Collect table data for the page:

```python
table = soup.find('table',{'class': 'dextable'})
table_data=table.find_all('td')
len(table_data)
```

There are 1344 physical pokemon moves as of generation 8.

```python
name = []
typ = []
cat = []
pp = []
power = []
accuracy = []
flavour = []
i = 0
for td in table_data:
    if i%7 == 0:
        name.append(td.text)
    if i%7 == 1:
        img = td.find('img')
        typ.append(img)
    if i%7 == 2:
        cat.append('physical')
    if i%7 == 3:
        pp.append(td.text)
    if i%7 == 4:
        power.append(td.text)
    if i%7 == 5:
        accuracy.append(td.text)
    if i%7 == 6:
        flavour.append(td.text)
    i = i+1
```

```python
physical = pd.DataFrame({'MOVE':name})
physical['TYPE'] = typ
physical['CATEGORY'] = cat
physical['PP'] = pp
physical['POWER'] = power
physical['ACCURACY'] = accuracy
physical['FLAVOUR'] = flavour
physical= physical.replace('\n','', regex=True)
physical = physical.replace('\t','', regex=True)
physical = physical.replace('\r','', regex=True)
physical = physical.iloc[1:]
physical['TYPE'] = physical['TYPE'].astype(str).replace('<img src="/attackdex/type/','',regex=True).replace('.gif"/>','',regex=True)
physical
```

# Merge dataframes

```python
frames = [physical,special, status]
df = pd.concat(frames)
df.dtypes
```

```python
df['ACCURACY'] = df['ACCURACY'].replace('--',101,regex=True)
#df['POWER'] = df['POWER'].replace('--',np.NaN)
#df['POWER'] = df['POWER'].values.astype(str).astype(int)
df['ACCURACY'] = df['ACCURACY'].values.astype(str).astype(int)
df['PP'] = df['PP'].values.astype(str).astype(int)
df = df.set_index('MOVE')
```
