# PRELIMINARIES

```python
import bs4 as bs
import urllib.request
import pickle
import pandas as pd
import numpy as np
#website to scrape
sauce = urllib.request.urlopen('https://www.serebii.net/pokedex-swsh/charizard/').read()
```

we use the package beautiful soup (bs4) for our web scraping. Based on the makeup of the website we can only scrape one pokemon's moves at one time using this method. Down the line this can be turned into one function to extract all a pokemon's moves.

```python
soup = bs.BeautifulSoup(sauce,'lxml')
```

# SCRAPE MOVES OF ONE POKEMON

```python
#collect all data tables
table = soup.find_all('table',{'class': 'dextable'})
```

Similarly to when we scraped the nationaldex, here we can acquire the table information from beautiful soup. However in this case the page we are scraping has multiple tables on it and so to get what we want (a dataframe with a list of all moves the pokemon can learn) we need to analyse the outputs of this page and our scraping. Using some experimenting it became clear that it is the 9th table on the page or the table with index 8 which is our starting point and we want to scrape all the data from this table and those below.

# Level Up Moves

```python
level = []
move = []
cat = []
typ = []
power = []
accuracy = []
pp = []
eff_percent = []
flavour = []
final = []
for t in table[8].find_all('tr'):
    for i in t.find_all('td'):
        trial = i
        final.append(trial)
i = 1
while i < len(final):
    if i%9 == 1:
        level.append(final[i].text)
    if i%9 == 2:
        move.append(final[i].text)
    if i%9 == 3:
        typ.append(final[i])
    if i%9 == 4:
        cat.append(final[i])
    if i%9 == 5:
        power.append(final[i].text)
    if i%9 == 6:
        accuracy.append(final[i].text)
    if i%9 == 7:
        pp.append(final[i].text)
    if i%9 == 8:
        eff_percent.append(final[i].text)
    if i%9 == 0:
        flavour.append(final[i].text)
    i = i+1
```

# remove regex characters in links

```python
result['TYPE'] = result['TYPE'].astype(str).replace('.*(?=/type/)','',regex=True).replace('/type/','',regex=True).replace('(?=.gif).*','',regex=True).replace('.gif','',regex=True)
result['CATEGORY'] = result['CATEGORY'].astype(str).replace('.*(?=/type/)','',regex=True).replace('/type/','',regex=True).replace('(?=.png).*','',regex=True).replace('.png','',regex=True)
result = result.drop_duplicates(subset=['MOVE'],keep='first') #since the rows are now sorted in descending order
result = result.set_index('MOVE')
result
```
